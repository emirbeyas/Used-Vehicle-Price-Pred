```python
selected_Fiyat = df['Fiyat']
std = selected_Fiyat.std()
avg = selected_Fiyat.mean()
three_sigma_plus = avg + (3*std)
three_sigma_minus = avg - (3*std)
print(three_sigma_plus, " ", three_sigma_minus)
for i in df[((selected_Fiyat > three_sigma_plus) | (selected_Fiyat < three_sigma_minus))].index:
    df.drop(i, inplace = True)
```

```python
df[['marka', 'Fiyat']].columns
```

```python
for column in df[[ 'Fiyat']].columns:
    for marka in df['marka'].unique():
        selected_marka = df[df['marka'] == marka]
        selected_column = selected_marka[column]
        
        q1 = selected_column.quantile(0.25)
        q3 = selected_column.quantile(0.75)
        
        iqr = q3 - q1
        
        min = q1 - (1.5 * iqr)
        max = q3 + (1.5 * iqr)
        
        print(column, marka, "| min= ", min, "max= ", max)
        
        maxidx = df[(df['marka'] == marka) & (df[column] > max)].index
        minidx = df[(df['marka'] == marka) & (df[column] < min)].index
        
        df.drop(index = maxidx, inplace = True )
        df.drop(index = minidx, inplace = True )
        
```

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split, GridSearchCV,cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df.drop(["Fiyat"], axis = 1)
y = df["Fiyat"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.27, random_state = 42)
```

```python
params = {"colsample_bytree":[0.4,0.5,0.6],
          "gamma":[0.1,0.2],
         "learning_rate":[0.01,0.02,0.05],
         "max_depth":[2,3,4,5],
         "n_estimators":[100,200,500,1000]}
```

```python
xgb = XGBRegressor()
```

```python
grid = GridSearchCV(xgb, params, cv =10, n_jobs = -1, verbose = 2)
```

```python
grid.fit(X_train, y_train)
```

```python
xgb1 = XGBRegressor(colsample_bytree = 0.5,learning_rate = 0.05, max_depth = 5, n_estimators = 1000)
```

```python
model_xgb = xgb1.fit(X_train, y_train)
```

```python
model_xgb.predict(X_test)[15:20]
```

```python
model_xgb.score(X_test, y_test)
```
